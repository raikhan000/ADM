```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
T = int(input())
for i in range(T):

    try:
        a,b = map(int,input().split())
        print(int(a/b))
    except ZeroDivisionError as e:
        print("Error Code: integer division or modulo by zero")
    except  ValueError as e:
        print("Error Code:",e)
```
