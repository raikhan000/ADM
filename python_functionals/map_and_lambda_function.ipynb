```python
cube = lambda x: x**3 # complete the lambda function 

def fibonacci(n):
    # return a list of fibonacci numbers
    n1 = 0
    n2 = 1
    sequence = []
    count = 0
    if n == 1:
        return [0]
    else:
        while count < n:
            sequence.append(n1)
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1
        return sequence
        
    

```
