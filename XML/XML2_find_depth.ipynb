```python
import xml.etree.ElementTree as etree

maxdepth = 0
def depth(elem, level):
    global maxdepth
    if level > maxdepth:
        maxdepth = level
    for x in elem:
        depth(x,level+1)
    if level == -1 and len(elem)!= 0:
        maxdepth +=1    
    return maxdepth
    # your code goes here

if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```
