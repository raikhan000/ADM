```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import Counter

if __name__ == '__main__':
    s = input()
    name_popularity = Counter(s)
    d = {}
    
    for i in name_popularity:
        if name_popularity[i] in d:
            d[name_popularity[i]].append(i)
        else:
            d[name_popularity[i]] = [i]
    
    max_num = max(d.keys())  
    cnt = 0  
    
    while cnt!= 3:   
        if len(d[max_num])>1:
            d[max_num].sort()
            for j in d[max_num]:
                if cnt == 3:
                    break
                print(j, max_num)
                cnt+=1
                
        else:
            print(*d[max_num], max_num)
            cnt+=1
        if len(d.keys())>1:
            d.pop(max_num)
            max_num = max(d.keys())
```
