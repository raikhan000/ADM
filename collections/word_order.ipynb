```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import OrderedDict

N = int(input())
strings = OrderedDict()
for i in range(N):
    a = input()
    if a in strings:
        strings[a]+=1
    else:
        strings[a] = 1

print(len(strings))
for i in strings:
    print(strings[i],end=" ")    
```
