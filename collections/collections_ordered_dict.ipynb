```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import OrderedDict
N = int(input())
items = OrderedDict()

for i in range(N):
    product_name = input().split()
    product_price  = product_name[len(product_name)-1]
    product_name.pop()
    product_name = " ".join(product_name)
    if product_name in items:
        items[product_name] += int(product_price)
    else:
        items[product_name] = int(product_price)
for i in items:
    print(i, items[i])
    
```
