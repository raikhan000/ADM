```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import deque

T = int(input())
Pyle = []
for i in range(T):
    n = int(input())
    sideLength = deque(map(int, input().split()))

    while len(sideLength)!= 1:
        a = sideLength.pop()
        b = sideLength.popleft()
        if a > b:
            Pyle.append(a)
            sideLength.appendleft(b)
        else:
            Pyle.append(b)
            sideLength.append(a)
    Pyle.append(sideLength.pop())
    if Pyle == sorted(Pyle,reverse= True):
        
        print("Yes")
    else:
        print("No")
    Pyle = []
    
```
