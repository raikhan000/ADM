```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import defaultdict

n, m = map(int, input().split())
A = defaultdict(list)
B = []
cnt = 1
for i in range(1, n+1):
    string = input()
    A[string].append(i)
    
for i in range(m):
    B.append(input())

for i in B:
    if i in A:
        print(*A[i])
    else:
        print(-1)
```
