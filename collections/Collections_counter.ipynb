```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

from collections import Counter

X = int(input())
shoe_sizes_list = list(map(int,input().split()))
shoe_number = Counter(shoe_sizes_list)
N = int(input())

prices = 0
for i in range(N):
    customer = list(map(int,input().split()))
    if customer[0] in shoe_number and shoe_number[customer[0]] >= 1:
        prices += customer[1]
        shoe_number[customer[0]] -= 1
print(prices)
        
            
```
