```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import namedtuple

N = int(input())

column_names = input()
Sheet = namedtuple("Sheet",column_names)

sum_marks=0
for i in range(N):
    string = input().split()
    student_data = Sheet(*string)
    sum_marks += int(student_data.MARKS)

print(sum_marks/N)
    
        
    
```
