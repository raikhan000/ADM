```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re
n = int(input())

pattern  = re.compile(r'^[a-z0-9A-Z]{10}$')

for i in range(n):
    s = input()
    m = pattern.match(s)
    if m:
        m1 = re.findall(r'\d', s)
        m2 = re.findall(r'[A-Z]', s)
        if len(m1)>= 3 and len(m2)>=2 and len(set(s)) == len(s): 
            print('Valid')
        else:
            print('Invalid')
    else:
        print('Invalid')
```
