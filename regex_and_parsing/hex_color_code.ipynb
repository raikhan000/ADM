```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())
pattern = re.compile(r'#([0-9a-f]{6})?([0-9a-f]{3})?(?=[;,)])',re.IGNORECASE)

for i in range(n):
    color = input()
    matches = pattern.finditer(color)
    for match in matches:
        print(match.group(0))
    
```
