```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

s = input()
k = input()
pattern = re.compile(r'(?=('+k+'))')
matches = pattern.finditer(s)
flag = False
for i in matches:
    flag = True
    print((i.start(1),i.end(1)-1))

if flag == False:
    print((-1,-1))
```
