```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re
n = int(input())

pattern = re.compile(r'^[4-6][0-9]{3}-?[0-9]{4}-?[0-9]{4}-?[0-9]{4}$')
pattern_2 = re.compile(r'0{4}|1{4}|2{4}|3{4}|4{4}|5{4}|6{4}|7{4}|8{4}|9{4}')

for i in range(n):
    s = input()
    m = pattern.match(s)
    if m:
        if '-' in s:
            s = s.split('-')
            s = ''.join(s)
        m2 = pattern_2.search(s)
        if m2:
            print('Invalid')
        else:
            print('Valid')
    else:
        print('Invalid')
```
