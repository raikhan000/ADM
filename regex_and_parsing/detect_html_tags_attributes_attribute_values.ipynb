```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(tag)
        for e in attrs:
            print("-> "+e[0]+" > "+str(e[1]))

    def handle_startendtag(self, tag, attrs):
        print(tag)
        for e in attrs: 
            print("-> "+e[0]+" > "+str(e[1]))
n = int(input())
parser = MyHTMLParser()
for i in range(n):
    s = input()
    parser.feed(s)
```
