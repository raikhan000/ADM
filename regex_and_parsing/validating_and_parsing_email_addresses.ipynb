```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re
n = int(input())
pattern1 = r"[ |<|>]"
pattern2 = re.compile(r"^[a-zA-Z][a-zA-Z0-9._-]+@([a-z])+\.([a-z]){1,3}$")
for i in range(n):
    s = input()
    a = re.split(pattern1, s)
    a = list(filter(None, a))
    matches = pattern2.match(a[1])
    if bool(matches):
        print(a[0]+" <"+a[1]+">")
```
