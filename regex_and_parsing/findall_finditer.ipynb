```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re
s = input()
pattern = re.compile(r'(?<=[bcdfghjklmnpqrstvwxz])[aeiou]{2,}(?=[bcdfghjklmnpqrstvwxz])', re.IGNORECASE)
matches = pattern.findall(s)
if matches:
    for match in matches:
        print(match)
else:
    print(-1)
```
