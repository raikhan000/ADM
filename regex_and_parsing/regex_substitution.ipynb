```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

n = int(input())
       
for i in range(n):
    s = input()
    pattern = re.compile(r'(?<=\s)(&{2}|\|{2})(?=\s)')
    m = pattern.findall(s)
    if m:
        for i in m:
            s = re.sub(r" && ", ' and ', s)
            s = re.sub(r" \|\| ", ' or ', s)
        print(s)
    
    else:
        print(s)
        
```
