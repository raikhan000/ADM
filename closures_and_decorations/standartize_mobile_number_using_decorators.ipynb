```python
def wrapper(f):
    def fun(l):
        # complete the function
        for i in range(len(l)):
            if len(l[i]) == 11:
                l[i] = '+91' +' ' + l[i][1:6] + ' ' + l[i][6:]
            elif len(l[i]) == 12:
                l[i] = '+91' +' ' + l[i][2:7] + ' ' + l[i][7:]
            elif len(l[i])==10:
                l[i] = '+91' +' ' + l[i][0:5] + ' ' + l[i][5:]
            elif l[i][0] == '+':
                l[i] = l[i][0:3] + ' ' +l[i][3:8] +' '+ l[i][8:]
        f(l)
    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 
```
