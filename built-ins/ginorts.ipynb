```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
def f(elem):
    if ord(elem)>=ord('a') and ord(elem)<=ord('z'):
        return ord(elem) - ord('a')
    if ord(elem)>=ord('A') and ord(elem)<=ord('Z'):
        return ord(elem)
    if ord(elem) >= ord('0') and ord(elem) <= ord('9'):
        if int(elem)%2!=0:
            return ord(elem)+ord('a')
        if int(elem)%2==0:
            return ord(elem)+ord('z')
            
a = input()
a = sorted(a,key=f)
print("".join(a))
```
