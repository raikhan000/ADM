```python
if __name__ == '__main__':
    s = input()
    flag_alphanumeric = False
    flag_alphabetic = False
    flag_digits = False
    flag_lowercase = False
    flag_uppercase = False
    for i in s:
        if i.isalnum():
            flag_alphanumeric = True
        if i.isalpha():
            flag_alphabetic = True
        if i.isdigit():
            flag_digits = True
        if i.islower():
            flag_lowercase = True
        if i.isupper():
            flag_uppercase = True
    print(flag_alphanumeric)
    print(flag_alphabetic)
    print(flag_digits)
    print(flag_lowercase)
    print(flag_uppercase)
            
```
