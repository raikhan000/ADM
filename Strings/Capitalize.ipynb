```python
# Complete the solve function below.
def solve(s):
    flag = False
    s1 = ""
    for i in range(len(s)):
        if s[i] == ' ':
            s1+=s[i]
            flag = False
            continue
        elif s[i].isalnum() and flag == False:
            flag = True
            s1 += s[i].upper()
        else:
            s1 += s[i]
    return s1

```
