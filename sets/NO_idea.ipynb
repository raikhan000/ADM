```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import Counter
n, m = input().split()
integers = input().split()

A = set(input().split())
B = set(input().split())

cnt = Counter(integers)
integer_set = set(integers)
happiness = 0

for i in integer_set&A:
    happiness+=cnt[i]
for i in integer_set&B:
    happiness-=cnt[i]



print(happiness)
```
