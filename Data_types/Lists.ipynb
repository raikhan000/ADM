```python
if __name__ == '__main__':
    commands = []
    a = []
    N = int(input())
    for i in range(N):
        commands.append(input().split())
    for i in commands:
        if i[0] == "insert":
            a.insert(eval(i[1]), eval(i[2]))
        elif i[0] == "print":
            print(a)
        elif i[0] == "remove":
            a.remove(eval(i[1]))
        elif i[0] == "append":
            a.append(eval(i[1]))
        elif i[0] == "sort":
            a.sort()
        elif i[0] == "pop":
            a.pop()
        elif i[0] == "reverse":
            a.reverse()
            
```
