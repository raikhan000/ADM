```python
if __name__ == '__main__':
    records = []
    def rec(elem):
        return elem[1]
    def rec1(elem):
        return elem[0]
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name, score])
    records.sort(key = rec)
    min_score = records[0][1]
    
    for i in range(len(records)):
        if(records[i][1] != min_score):
            min_score = records[i][1]
            break
    records.sort(key = rec1)
    
    for i in range(len(records)):
        if(records[i][1] == min_score):
            print(records[i][0])
    
        
```
